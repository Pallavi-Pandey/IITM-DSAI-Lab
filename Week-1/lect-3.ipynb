- Curve 2: y = 5-3x for x in range (-5, 5) and add a title to the plot.

```python
x = np.linspace(0, 1, 20)
y = 5 - 3 * x
plt.plot(x, y)
plt.title('y = 5 - 3x');
```

- Curve 3: y = x^2
    - 5, 10, 20, 50 equally spaced points in the interval

```python
x = np.linspace(-1, 1, 50)
y = x ** 2
# plt.scatter(x, y)
plt.plot(x, y)
plt.title('$y = x^{2}$')
plt.axhline(color = 'black',
            linestyle = '--',
            linewidth = 0.8)
plt.axvline(color = "red",
            linestyle = '--',
            linewidth = 0.8);
```

- Curve 4: y = 3 and x = 5 on same plot.
    - first one blue
    - second one red
    - limit x in range (0, 6)
    - limit y in range (0, 6)
    - label x and y axis
    - add a title
    - add a legend

```python
plt.axhline(y=3, color='blue', label='y = 3')
plt.axvline(x=5, color='red', label='x = 5')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Horizontal and Vertical Lines')
plt.legend()
plt.show()
```

Subplot



```python
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
plt.subplot(2, 2, 1)
x = np.linspace(1 / np.e ** 2, np.e ** 3)
y = np.log(x)
plt.plot(x, y)
plt.title(r'$y = \ln(x)$')

plt.subplot(2, 2, 2)
x = np.linspace(0, 3)
y = np.exp(x)
plt.plot(x, y)
plt.title(r'$y = e^{x}$')

plt.subplot(2, 2, 3)
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)
plt.title(r'$y = \sin(x)$')

plt.subplot(2, 2, 4)
x = np.linspace(1, 5)
y = 1 / x
plt.plot(x, y)
plt.title(r'$y = \frac{1}{x}$');
```
